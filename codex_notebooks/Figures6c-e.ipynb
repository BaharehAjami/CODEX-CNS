```python
# Load necessary libraries
import sys
import os
import anndata as ad
import pandas as pd
import scanpy as sc
import seaborn as sns; sns.set(color_codes=True)
import collections
#import scimap as sm

import matplotlib.pyplot as plt
#import squidpy as sq

import warnings
warnings.filterwarnings('ignore')
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor='white')
```

```python
sc.logging.print_header()
```

```python



adata= sc.read_h5ad('fabian_metadata_allcells.h5ad')

adata = adata[adata.obs['Parent'].isin(['Grey Matter', 'White Matter'])]
adata = adata[~adata.obs['phenotype_translated'].isin(['dense_aBeta', 'diffuse_aBeta'])]

```

```python
srs = adata.obs.groupby(['ImageID','Parent'])['phenotype_translated'].value_counts()

srs.index.names = ['ImageID', 'Parent', 'phenotype_translated']
```

```python

num = adata.obs.groupby(['ImageID','Parent'])['phenotype_translated'].value_counts()
num.index.names = ['ImageID', 'Parent', 'phenotype_translated']

denom = adata.obs.groupby(['ImageID','Parent']).size()

statsDF = 100 * (num/denom)
statsDF = statsDF.reset_index()
statsDF.columns = ['ImageID','Parent','phenotype_translated','percentage']
```

```python
#Map 'Parent' to a new column 'Region' with the abbreviations
nameMap = {'Grey Matter':'GM','White Matter':'WM'}
statsDF['Region'] = statsDF.Parent.map(nameMap)
```

```python
set(adata.obs.phenotype_translated)
```

```python
# assign colors to phenotypes
cMapDict={'Astrocytes': '#FDD835',
 'Microglia': '#00DBF2',
 'Neurons': '#4CAF50',
 'Oligodendrocytes': '#FF00FF'}
```

```python
# performing only on the following groups...


groups_with_olig2 = list(statsDF[(statsDF['percentage'] > 0) & (statsDF['phenotype_translated'] == 'Oligodendrocytes')]['ImageID'].unique())

statsDF = statsDF[statsDF.ImageID.isin(groups_with_olig2)]

# plot cell type stats for Ctl-1 sample
sns.set_style('white')

fig,ax=plt.subplots(1,1,figsize=(5,5))
sns.barplot(data=statsDF[statsDF.ImageID=='Ctl-1'], x='Region', y ='percentage',hue='phenotype_translated',orient='v',ax=ax,
            hue_order=['Astrocytes', 'Microglia', 'Neurons', 'Oligodendrocytes'],
              palette=cMapDict)
ax.set_ylabel('% of cells')
ax.set_xlabel('')
ax.set_yticklabels(['0%','10%','20%','30%','40%','50%','60%','70%'])
plt.legend(title='',facecolor=None)
plt.tight_layout()
plt.savefig('figures/GM_WM_proportion_barchart.svg', format = 'svg', dpi = 300)
plt.savefig('figures/GM_WM_proportion_barchart.png', format = 'png', dpi = 300)




sns.set_style('white')

fig,ax=plt.subplots(1,1,figsize=(5,5))
sns.barplot(data=statsDF[statsDF.ImageID.isin(groups_with_olig2)], x='Region', y ='percentage',hue='phenotype_translated',orient='v',ax=ax,
            hue_order=['Astrocytes', 'Microglia', 'Neurons', 'Oligodendrocytes'],
              palette=cMapDict, capsize=0.1)
sns.stripplot(data=statsDF[statsDF.ImageID.isin(groups_with_olig2)], x='Region', y ='percentage',hue='phenotype_translated',ax=ax,
            hue_order=['Astrocytes', 'Microglia', 'Neurons', 'Oligodendrocytes'],
            dodge = True, legend=False, color='black')

ax.set_ylabel('% of cells')
ax.set_xlabel('')
ax.set_yticklabels(['0%','10%','20%','30%','40%','50%','60%','70%'])
plt.legend(title='',facecolor=None)
plt.tight_layout()
plt.savefig('figures/GM_WM_proportion_barchart_all.svg', format = 'svg', dpi = 300)
plt.savefig('figures/GM_WM_proportion_barchart_all.png', format = 'png', dpi = 300)

print(statsDF.groupby(['Parent', 'phenotype_translated']).mean())



sns.set_style('white')

fig,ax=plt.subplots(1,1,figsize=(5,5))
sns.barplot(data=statsDF[statsDF.ImageID.isin(['AD-1', 'AD-2', 'AD-3', 'AD-4'])], x='Region', y ='percentage',hue='phenotype_translated',orient='v',ax=ax,
            hue_order=['Astrocytes', 'Microglia', 'Neurons', 'Oligodendrocytes'],
              palette=cMapDict, capsize=0.1)
sns.stripplot(data=statsDF[statsDF.ImageID.isin(['AD-1', 'AD-2', 'AD-3', 'AD-4'])], x='Region', y ='percentage',hue='phenotype_translated',ax=ax,
            hue_order=['Astrocytes', 'Microglia', 'Neurons', 'Oligodendrocytes'],
            dodge = True, legend=False, color='black')

ax.set_ylabel('% of cells')
ax.set_xlabel('')
ax.set_yticklabels(['0%','10%','20%','30%','40%','50%','60%','70%'])
plt.legend(title='',facecolor=None)
plt.tight_layout()
plt.savefig('figures/GM_WM_proportion_barchart_ad.svg', format = 'svg', dpi = 300)
plt.savefig('figures/GM_WM_proportion_barchart_ad.png', format = 'png', dpi = 300)

print(statsDF.groupby(['Parent', 'phenotype_translated']).mean())




sns.set_style('white')

fig,ax=plt.subplots(1,1,figsize=(5,5))
sns.barplot(data=statsDF[statsDF.ImageID.isin(['Ctl-1', 'Ctl-2', 'Ctl-3', 'Ctl-4'])], x='Region', y ='percentage',hue='phenotype_translated',orient='v',ax=ax,
            hue_order=['Astrocytes', 'Microglia', 'Neurons', 'Oligodendrocytes'],
              palette=cMapDict, capsize=0.1)
sns.stripplot(data=statsDF[statsDF.ImageID.isin(['Ctl-1', 'Ctl-2', 'Ctl-3', 'Ctl-4'])], x='Region', y ='percentage',hue='phenotype_translated',ax=ax,
            hue_order=['Astrocytes', 'Microglia', 'Neurons', 'Oligodendrocytes'],
            dodge = True, legend=False, color='black')

ax.set_ylabel('% of cells')
ax.set_xlabel('')
ax.set_yticklabels(['0%','10%','20%','30%','40%','50%','60%','70%'])
plt.legend(title='',facecolor=None)
plt.tight_layout()
plt.savefig('figures/GM_WM_proportion_barchart_control.svg', format = 'svg', dpi = 300)
plt.savefig('figures/GM_WM_proportion_barchart_control.png', format = 'png', dpi = 300)

print(statsDF.groupby(['Parent', 'phenotype_translated']).mean())
```

### Plot spatial map

```python


for image in adata.obs.ImageID.unique():
    fig,ax=plt.subplots(1,1,figsize=(10,8))
    sns.set_style('white')
    ss = adata[adata.obs.ImageID==image]

    sns.scatterplot(data=ss.obs,x='spatial_X',y='spatial_Y',hue='phenotype_translated', 
                    palette=cMapDict, s=2)
    plot_spat = plt.gca()
    ax.invert_yaxis()
    plt.axis('off')
    ax.legend(title='')
    plt.tight_layout()
    plt.savefig('figures/' + image + 'celltype_scatterplot.svg', format = 'svg', dpi = 300)
    plt.savefig('figures/' + image + 'celltype_scatterplot.png', format = 'png', dpi = 300)
```
